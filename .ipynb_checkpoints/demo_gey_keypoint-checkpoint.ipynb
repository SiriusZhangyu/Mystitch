```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
```

```python
curve_img = cv2.imread('binary.png', 0)
curve_img = 255 - curve_img
plt.imshow(curve_img, cmap='gray')
```

```python
binary_img = np.zeros(curve_img.shape, np.uint8)
binary_img[curve_img < 128] = 0
binary_img[curve_img >= 128] = 1
skel_bool = skeletonize(binary_img)
skel_img = np.zeros(curve_img.shape, np.uint8)
skel_img[skel_bool==True] = 255
plt.imshow(skel_img, cmap='gray')
```

```python
# In each grid, select one skeleton pixel
grid_size = 30
skel_point_list = list()
for x in range(0, skel_img.shape[0]-grid_size, grid_size):
    for y in range(0, skel_img.shape[1]-grid_size, grid_size):
        grid_img = skel_img[y:y+grid_size, x:x+grid_size]
        grid_center = np.array([grid_size/2, grid_size/2])
        skel_points = np.argwhere(grid_img == 255)
        if skel_points.shape[0] > 0:
            skel_point_list.append(skel_points[np.argmin(np.sum((skel_points - grid_center)**2, axis=1))] + np.array([y,x]))
```

```python
# Show selected pixel
vis_skel_img = cv2.cvtColor(skel_img, cv2.COLOR_GRAY2RGB)
for p in skel_point_list:
    [py, px] = p
    cv2.circle(vis_skel_img, (px, py), 3, (255,0,0), -1)
plt.imshow(vis_skel_img, cmap='gray')
```
